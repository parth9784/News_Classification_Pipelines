```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
```

```python
def preprocess(news):
    tokens = word_tokenize(news)
    sent=[]
    for i in tokens:
        if(i.lower() not in stop_words and i not in punctuation):
            sent.append(i.lower())
    return " ".join(sent)        
```

```python
df_final=df_final["text"].apply(preprocess)
```

```python
df_prepro=pd.concat([df_final,df_istrue],axis=1)
```

```python
df_prepro.shape
```

```python
df_prepro.to_csv("dfprepro.csv")
```

First By converting the preprocessed news into vectors By using Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train,X_test,y_train,y_test=train_test_split(df_prepro.text,df_prepro.istrue,test_size=0.2,random_state=1210,stratify=df_prepro.istrue)
```

```python
X_train
```

```python
y_train
```

```python
clf1=Pipeline([
    ("Countvectorizer",CountVectorizer()),
    ("classifier",RandomForestRegressor())
])
```

```python
clf1
```

```python
clf1.fit(X_train,y_train)
```

```python
import joblib
joblib.dump(clf1,"Clf1.pkl")
```

```python
y_pred1=clf1.predict(X_test)
```

```python
y_Test=y_test.values
```

```python
y_predint=y_pred1.astype(int)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
accuracy1 = accuracy_score(y_Test, y_predint)
precision1=precision_score(y_Test,y_predint)
print("Accuracy is: ",accuracy1)
print("Precision Score: ",precision1)
```

```python
confusion_matrix(y_Test,y_predint)
```

Second By vectorization using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
clf2=Pipeline([
    ("tf-idf",TfidfVectorizer()),
    ("RandomRegressor",RandomForestRegressor())
])
```

```python
clf2
```

```python
clf2.fit(X_train,y_train)
```

```python
joblib.dump(clf2,"Clf2.pkl")
```

```python
y_pred2=clf2.predict(X_test)
```

```python
y_predint2=y_pred2.astype(int)
```

```python
accuracy2 = accuracy_score(y_Test, y_predint2)
precision2=precision_score(y_Test,y_predint2)
print("Accuracy is: ",accuracy2)
print("Precision Score: ",precision2)
```

```python
confusion_matrix(y_Test,y_predint2)
```
